```python
num_cols = ["Spotify Streams", "Spotify Playlist Count", "Spotify Playlist Reach", "YouTube Views", "YouTube Likes", "TikTok Likes", "TikTok Posts", "TikTok Views",
"Soundcloud Streams", "SiriusXM Spins", "Shazam Counts", "Spotify Popularity", "YouTube Playlist Reach", "Apple Music Playlist Count", "AirPlay Spins",
"Deezer Playlist Count", "Deezer Playlist Reach",  "Amazon Playlist Count", "Pandora Streams", "Pandora Track Stations"]

def convert_to_numeric(df, col_list):

    df_clean = df.copy()

    for col in col_list:
        try:
            df_clean[col] = pd.to_numeric(
                df_clean[col].astype(str).str.replace(",", "", regex=False),
                errors='coerce'
            ).fillna(0).astype('int')

        except ValueError as e:
            print(f"Could not convert column {col} to int: {e}")

    return df_clean

def convert_to_datetime(df, col_list):

    df_clean = df.copy()
    
    for col in col_list:
        try:
            df_clean[col] = pd.to_datetime(df_clean[col], errors='coerce')


            df_clean['Release Year'] = df_clean[col].dt.year
            df_clean['Days Since Release'] = (pd.Timestamp.now() - df_clean[col]).dt.days
            df_clean['Track Age Category'] = pd.cut(df_clean['Release Year'],
                                                    bins=[0, 2021, 2023, 2025],
                                                    labels=['Pre (2022)', 'Recent (2022-2023)', 'New (2024+)'])

        except ValueError as e:
            print(f"Could not convert column {col} to datetime: {e}")

    return df_clean

df_clean = convert_to_numeric(df, num_cols)
df_clean = convert_to_datetime(df_clean, ["Release Date"])
df_clean = df_clean.drop(columns=["TIDAL Popularity"])
df_clean.info()
```

Cross Platform Analysis:

  1. Which Platforms have the strongest correlation with Spotify streams? (YouTube, TikTok, Apple Music)

  2. Do Explicit Tracks perform differently than clean tracks across platforms

  3. Analyse temporal patterns (release data trends)

  4. Visulations for cross-platform dynamics

```python
# Creating a function to find the columns that have the strongest correlation with Spotify Streams using spearman scoring

def return_col_with_highest_coeff(df, cols):
    
    from scipy.stats import spearmanr

    df_filtered = df[df['Spotify Streams'] > 0].copy()
    corrs = []

    for col in cols:
        temp_df = df_filtered[df_filtered[col] > 0]
        corr, _ = spearmanr(temp_df['Spotify Streams'], temp_df[col])
        corrs.append((col, corr))

    corrs.sort(key=lambda x: abs(x[1]), reverse=True)
    top_3_corrs = corrs[:5]

    return top_3_corrs

# Return columns that are NOT Spotify that have the highest correlation with Spotify

cols =  ["YouTube Views", "YouTube Likes", "TikTok Likes", "TikTok Posts", "TikTok Views",
"Soundcloud Streams", "SiriusXM Spins", "Shazam Counts", "Spotify Popularity", "YouTube Playlist Reach", "Apple Music Playlist Count", "AirPlay Spins",
"Deezer Playlist Count", "Deezer Playlist Reach",  "Amazon Playlist Count", "Pandora Streams", "Pandora Track Stations"]

cols_with_highest = return_col_with_highest_coeff(df_clean, cols)
print(cols_with_highest)
```

```python
def plot_relationship(df, x=None, y=None):
    for c in y:
        col = c[0]
        corr_value = c[1]
        df_temps = df[(df[col] > 0) & (df['Spotify Streams'] > 0)].copy()
        
        # Create a scatter plot
        plt.figure(figsize=(10, 6))
        sns.scatterplot(x='Spotify Streams', y=col, 
                        data=df_temps, 
                        alpha=0.5,
                        hue="Explicit Track",
                        s=40)
        
        # Set log scale for axes
        plt.xscale('log')
        plt.yscale('log')
        
        # Add title and labels
        plt.title(f"Relationship between Spotify Streams & {col}: {corr_value:.3f}", fontsize=14)
        plt.xlabel('Spotify Streams (log scale)', fontsize=12)
        plt.ylabel(f"{col} (log scale)", fontsize=12)
        
        # Improve layout
        plt.grid(True, alpha=0.5)
        plt.tight_layout()
        plt.show()

plot_relationship(df_clean, x='Spotify Streams', y=cols_with_highest)
```
