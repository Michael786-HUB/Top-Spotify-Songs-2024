Cleaning the dataset:

Many of the variables have the wrong the data tytpe and are stored as objects. Likely due to the numerical values having string characters (commas). Create a function convert_to_numerical that targets specific columns and turns into a numerical volumne

```python
num_cols = ["Spotify Streams", "Spotify Playlist Count", "Spotify Playlist Reach", "YouTube Views", "YouTube Likes", "TikTok Likes", "TikTok Posts", "TikTok Views",
"Soundcloud Streams", "SiriusXM Spins", "Shazam Counts", "Spotify Popularity", "YouTube Playlist Reach", "Apple Music Playlist Count", "AirPlay Spins",
"Deezer Playlist Count", "Deezer Playlist Reach",  "Amazon Playlist Count", "Pandora Streams", "Pandora Track Stations"]

def convert_to_numeric(df, col_list):

    df_clean = df.copy()

    for col in col_list:
        try:
            df_clean[col] = pd.to_numeric(
                df_clean[col].astype(str).str.replace(",", "", regex=False),
                errors='coerce'
            ).fillna(0).astype('int')

        except ValueError as e:
            print(f"Could not convert column {col} to int: {e}")

    return df_clean

df_clean = convert_to_numeric(df, num_cols)
df_clean.info()
```

Can use scatterplots to plot the relationship between two variables.

the log scale attribute scales variables with different value ranges 

```python
from scipy.stats import spearmanr

df_new = df_clean[(df_clean['YouTube Views'] > 0) & (df_clean['Spotify Streams'] > 0)]

# Calculate correlation
corr, p_value = spearmanr(df_new['Spotify Streams'], df_new['YouTube Views'])

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Spotify Streams', y='YouTube Views', data=df_new, 
                alpha=0.5, s=40, hue='Explicit Track')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Spotify Streams (log scale)', fontsize=12)
plt.ylabel('YouTube Views (log scale)', fontsize=12)
plt.title(f'Spotify Streams vs YouTube Views\nSpearman Correlation: {corr:.3f}', 
          fontsize=14)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"Correlation: {corr:.3f}")
```

```python
def return_col_with_highest_coeff(df, cols):
    
    df_filtered = df[df['Spotify Streams'] > 0].copy()
    corrs = []

    for col in cols:
        temp_df = df_filtered[df_filtered[col] > 0]
        corr, _ = spearmanr(temp_df['Spotify Streams'], temp_df[col])
        corrs.append((col, corr))

    corrs.sort(key=lambda x: abs(x[1]), reverse=True)
    top_3_corrs = corrs[:5]

    return top_3_corrs

cols =  ["Spotify Playlist Count", "Spotify Playlist Reach", "YouTube Views", "YouTube Likes", "TikTok Likes", "TikTok Posts", "TikTok Views"]
cols_with_highest = return_col_with_highest_coeff(df_clean, cols)
```

```python
def plot_relationship(df, x=None, y=None):
    for c in y:
        col = c[0]
        corr_value = c[1]
        df_temps = df[(df[col] > 0) & (df['Spotify Streams'] > 0)].copy()
        
        # Create a scatter plot
        plt.figure(figsize=(10, 6))
        sns.scatterplot(x='Spotify Streams', y=col, 
                        data=df_temps, 
                        alpha=0.5,
                        hue="Explicit Track",
                        s=40)
        
        # Set log scale for axes
        plt.xscale('log')
        plt.yscale('log')
        
        # Add title and labels
        plt.title(f"Relationship between Spotify Streams & {col}: {corr_value:.3f}", fontsize=14)
        plt.xlabel('Spotify Streams (log scale)', fontsize=12)
        plt.ylabel(f"{col} (log scale)", fontsize=12)
        
        # Improve layout
        plt.grid(True, alpha=0.5)
        plt.tight_layout()
        plt.show()

plot_relationship(df_clean, x='Spotify Streams', y=cols_with_highest)
```

Pearson correlation coefficent
  - Measures the linear relationships
  - Uses actual values
  - Is sensitive to outliers
  - Assumes normal distribution

Spearman
 - measures monotic relationships (consistent co)
 - Uses Rank Order
 - Robust against outliers

Linear relationships are where variables increase or decrease concurrently at a constant rate (so there is a positive or negative linear relationships) --- plotted on a graph, they would be a straight line.

In a monotonic relationship, the variables mvoe in the same relative direction, but they don't move at a constant rate.
